```python
data=pd.read_csv(r'segmentationdata.csv',index_col=0)
data.head()
```

```python
data.corr()
```

```python
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
plt.figure(figsize=(15,9))
s=sns.heatmap(data.corr(),
             annot=True,
             cmap='RdBu',
             vmin=-1,
             vmax=1)
s.set_yticklabels(s.get_yticklabels(),rotation=0,fontsize=12)
s.set_xticklabels(s.get_xticklabels(),rotation=0,fontsize=12)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize = (12, 9))
plt.scatter(data.iloc[:, 2], data.iloc[:, 4])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Visualization of raw data')
```

```python
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(data)
```

```python
segmentation_std
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(segmentation_std)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss,marker='o',linestyle='--')
plt.xlabel('number of clustes')
plt.ylabel('WCSS')
plt.title('K-means Clutering')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
kmeans.fit(segmentation_std)
```

```python
df_seg_means=data.copy()
df_seg_means['Segment k-means']=kmeans.labels_
```

```python
temp=df_seg_means.groupby(['Segment k-means']).mean()
temp
```

```python
temp.index
```

```python
temp['N Obs'] = df_seg_means[['Segment k-means','Sex']].groupby(['Segment k-means']).count()
temp['Prop Obs'] = temp['N Obs'] / temp['N Obs'].sum()
```

```python
temp
```

```python
df_seg_means['labels']=df_seg_means['Segment k-means'].map({0:'well-off',
                                                            1:'fewer oppor',
                                                            2:'standard',
                                                            3:'career focused'})
```

```python
df_seg_means.head()
```

```python
x_axis=df_seg_means['Age']
y_axis=df_seg_means['Income']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis,y_axis,hue=df_seg_means['labels'],palette=['g','r','c','m'])
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(12,9))
plt.plot(range(1,8),pca.explained_variance_ratio_.cumsum(),marker='o',linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
pca=PCA(n_components=3)
pca.fit(segmentation_std)
```

```python
pca.components_
```

```python
data_pca=pd.DataFrame(data=pca.components_,
                     columns=data.columns.values,
                     index=['Component 1','Component 2','Component 3'])
data_pca
```

```python
sns.heatmap(data_pca,
           vmin=-1,
           vmax=1,
           cmap='RdBu',
           annot=True)
plt.yticks([0,1,2],
           ['Component 1','Component 2','Component 3'],
           rotation=45,
           fontsize=9)
```

```python
score_pca=pca.transform(segmentation_std)
```

```python
score_pca
```

```python
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(score_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA Clustering')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans_pca.fit(score_pca)
```

```python
df_segm_pca_kmeans = pd.concat([data.reset_index(drop = True), pd.DataFrame(score_pca)], axis = 1)
df_segm_pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']
# The last column we add contains the pca k-means clustering labels.
df_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans
```

```python
df_segm_pca_kmeans_freq = df_segm_pca_kmeans.groupby(['Segment K-means PCA']).mean()
df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans_freq['N Obs'] = df_segm_pca_kmeans[['Segment K-means PCA','Sex']].groupby(['Segment K-means PCA']).count()
df_segm_pca_kmeans_freq['Prop Obs'] = df_segm_pca_kmeans_freq['N Obs'] / df_segm_pca_kmeans_freq['N Obs'].sum()
df_segm_pca_kmeans_freq = df_segm_pca_kmeans_freq.rename({0:'standard', 
                                                          1:'career focused',
                                                          2:'fewer opportunities', 
                                                          3:'well-off'})
df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans['Legend'] = df_segm_pca_kmeans['Segment K-means PCA'].map({0:'standard', 
                                                          1:'career focused',
                                                          2:'fewer opportunities', 
                                                          3:'well-off'})
```

```python
x_axis = df_segm_pca_kmeans['Component 2']
y_axis = df_segm_pca_kmeans['Component 1']
plt.figure(figsize = (10, 8))
sns.scatterplot(x_axis, y_axis, hue = df_segm_pca_kmeans['Legend'], palette = ['g', 'r', 'c', 'm'])
plt.title('Clusters by PCA Components')
plt.show()
```
